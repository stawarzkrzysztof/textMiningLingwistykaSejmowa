```python
from pathlib import Path
import json

IN_DIR = Path("/Users/krzysztofstawarz/GithubRepositories/textMiningLingwistykaSejmowa/1dataDownload/data/raw/term10/mp/scrapped_details")          
OUT_FILE = Path("/Users/krzysztofstawarz/GithubRepositories/textMiningLingwistykaSejmowa/1dataDownload/data/raw/term10/mp/scrapped_details_list.json")

items: list[dict] = []

files = sorted(IN_DIR.glob("*.json"), key=lambda p: int(p.stem))

for p in files:
    with p.open("r", encoding="utf-8") as f:
        obj = json.load(f)
        if not isinstance(obj, dict):
            raise ValueError(f"{p.name}: oczekiwano dict, jest {type(obj)}")
        items.append(obj)

OUT_FILE.parent.mkdir(parents=True, exist_ok=True)
with OUT_FILE.open("w", encoding="utf-8") as f:
    json.dump(items, f, ensure_ascii=False, indent=2)
```

```python
s = mpsClean["ukonczonaSzkola"].fillna("")

mpsClean["najwyzszyStopienNaukowy"] = np.select(
    [
        s.str.contains(r"-\s*profesor", case=False, regex=True),
        s.str.contains(r"-\s*doktor", case=False, regex=True),
        s.str.contains(r"-\s*magister\s+inżynier", case=False, regex=True),
        s.str.contains(r"-\s*magister", case=False, regex=True),
        s.str.contains(r"-\s*lekarz", case=False, regex=True),
        s.str.contains(r"-\s*inżynier", case=False, regex=True),
        s.str.contains(r"-\s*licencjat", case=False, regex=True),
    ],
    [
        # zbitki od "profesor..." / "doktor..." / "lekarz..." wyciągamy w całości do ( lub separatora
        s.str.extract(r"-\s*(profesor[^,;()]*)\s*(?=\(|$|,|;)", expand=False),
        s.str.extract(r"-\s*(doktor[^,;()]*)\s*(?=\(|$|,|;)", expand=False),
        "magister inżynier",
        "magister",
        s.str.extract(r"-\s*(lekarz[^,;()]*)\s*(?=\(|$|,|;)", expand=False),
        "inżynier",
        "licencjat",
    ],
    default=np.nan
)

mpsClean["najwyzszyStopienNaukowy"] = (
    mpsClean["najwyzszyStopienNaukowy"]
      .astype("string")
      .str.replace(r"\s+", " ", regex=True)
      .str.strip()
)
```
