```python
import re


ROMAN_RE = re.compile(r"\b(?:I|II|III|IV|V|VI|VII|VIII|IX|X)\b")
SEJM_SEG_RE = re.compile(r"Sejm RP\s+(.+?)\s+kadencji", re.IGNORECASE)

def count_prev_sejm_terms(staz: str) -> int:
    if pd.isna(staz):
        return 0
    s = str(staz).strip()
    if s.lower() == "brak":
        return 0

    m = SEJM_SEG_RE.search(s)
    seg = m.group(1) if m else s
    return len(ROMAN_RE.findall(seg))

mpsClean["ktoraKadencja"] = mpsClean["stazParlamentarny"].apply(count_prev_sejm_terms) + 1  # 'brak' => 1

mpsClean.head()
```

```python
import re
import pandas as pd


# ---------------------------
# Helpers
# ---------------------------
def _norm(x) -> str:
    if pd.isna(x):
        return ""
    return str(x).strip()


def _to_bool(x) -> bool:
    if isinstance(x, bool):
        return x
    s = _norm(x).lower()
    return s in {"true", "1", "t", "yes", "y"}


# ---------------------------
# Binning rules (order matters)
# ---------------------------
COMMENT_RULES = [
    # EU Parliament / MEP mandate
    (
        r"\bparlamentu europejskiego\b|mandat posła do parlamentu europejskiego",
        "Mandat w Parlamencie Europejskim",
        "Parlament Europejski",
    ),

    # Loss of eligibility
    (
        r"utrata prawa wybieralno",
        "Utrata prawa wybieralności",
        "Utrata prawa wybieralności",
    ),

    # Death
    (
        r"\bzmarł\b|\bzmarla\b|\bzmarła\b|\bzgon\b",
        "Zgon",
        "Zgon",
    ),

    # Local government
    (
        r"prezydenta miasta|burmistrza|wójta|wojta|marszałka województwa|marszalka wojewodztwa",
        "Objęcie funkcji w samorządzie",
        "Samorząd",
    ),

    # Central administration / high offices (sub-bins)
    (
        r"kancelarii prezydenta rp|gabinetu prezydenta rp|szefa kancelarii prezydenta rp|kancelarii prezydenta",
        "Objęcie funkcji w administracji centralnej",
        "Kancelaria Prezydenta RP",
    ),
    (
        r"sekretarza stanu",
        "Objęcie funkcji w administracji centralnej",
        "Sekretarz stanu",
    ),
    (
        r"narodowego banku polskiego|\bnbp\b",
        "Objęcie funkcji w administracji centralnej",
        "NBP",
    ),
]


def bin_comment(comment: str):
    c = _norm(comment)
    if not c:
        return (None, None)

    t = c.lower()
    for pattern, bin_label, subbin_label in COMMENT_RULES:
        if re.search(pattern, t, flags=re.IGNORECASE):
            return (bin_label, subbin_label)

    return ("Inne (z komentarza)", "Inne (z komentarza)")


def bin_reason(reason: str):
    r = _norm(reason)
    if not r:
        return None

    t = r.lower()
    if "zgon" in t:
        return "Zgon"
    if "zrzeczenie" in t:
        return "Zrzeczenie się mandatu"
    return "Inne (z powodu)"


def categorize_inactivity_row(row: pd.Series) -> pd.Series:
    """
    Tworzy:
    - powod_bin: z 'powodNieaktywnosci'
    - komentarz_bin + komentarz_subbin: z 'nieaktywnoscKomentarz'
    - nieaktywnosc_bin: finalny bin (preferuje komentarz, potem powód)
    - nieaktywnosc_subbin: sub-kategoria (jeśli jest)
    """
    active = _to_bool(row.get("czyAktywny"))

    if active:
        return pd.Series({
            "powod_bin": None,
            "komentarz_bin": None,
            "komentarz_subbin": None,
            "nieaktywnosc_bin": "Aktywny",
            "nieaktywnosc_subbin": None,
        })

    rbin = bin_reason(row.get("powodNieaktywnosci"))
    kbin, ksub = bin_comment(row.get("nieaktywnoscKomentarz"))

    final_bin = kbin or rbin or "Nieznane"
    final_sub = ksub  # subbin tylko z komentarza

    return pd.Series({
        "powod_bin": rbin,
        "komentarz_bin": kbin,
        "komentarz_subbin": ksub,
        "nieaktywnosc_bin": final_bin,
        "nieaktywnosc_subbin": final_sub,
    })


# ---------------------------
# Usage
# ---------------------------
mpsClean["czyAktywny"] = mpsClean["czyAktywny"].apply(_to_bool)

bins = mpsClean.apply(categorize_inactivity_row, axis=1)
mpsClean = pd.concat([mpsClean, bins], axis=1)

inactive = mpsClean[mpsClean["czyAktywny"] == False]

print("\nRozkład binów (nieaktywni):")
print(inactive["nieaktywnosc_bin"].value_counts(dropna=False))

print("\nRozkład bin + subbin (nieaktywni):")
print(
    inactive.groupby(["nieaktywnosc_bin", "nieaktywnosc_subbin"])
    .size()
    .sort_values(ascending=False)
)
```
